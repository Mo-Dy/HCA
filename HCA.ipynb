# **Clustering Analysis:**
# Hierarchical Clustering









```python
data = df.iloc[:,3:5]
```

```python
data.shape
```

```python
data.info()
```

### Dendrograms:


```python
import scipy.cluster.hierarchy as sch
dendro = sch.dendrogram(sch.linkage(data, method='ward'))
plt.title('Dendrogram')
plt.xlabel('Custmers')
plt.ylabel('Euclidean Distance')
plt.show()
#plt.axhline(y=200, color = 'r', linestyle='--')
```

### Agglomerative Heirarchical Algorithm:


```python
from sklearn.cluster import AgglomerativeClustering
clusters = AgglomerativeClustering(n_clusters =5, affinity ='euclidean', linkage = 'ward')
y = clusters.fit_predict(data)
```

```python
plt.scatter(data[:, 0], data[:, 1], c=clusters.labels_ )
plt.title('CUstomer Segmentation')
plt.xlabel('Annual Income(k$)')
plt.ylabel('Spending Score(1-100)')
```

### Define Customer's Segment of the Mall:

```python

plt.scatter(data[y == 0, 0], data[y == 0, 1], c = 'yellow', label = 'Careful')
plt.scatter(data[y == 1, 0], data[y == 1, 1], c = 'cyan', label = 'Standard')
plt.scatter(data[y == 2, 0], data[y == 2, 1],  c = 'green', label = 'Target')
plt.scatter(data[y == 3, 0], data[y == 3, 1] , c = 'red', label = 'Careless')
plt.scatter(data[y == 4, 0], data[y == 4, 1], c = 'orange', label = 'Sensible')

plt.title('CUstomer Segmentation')
plt.xlabel('Annual Income(k$)')
plt.ylabel('Spending Score(1-100')
plt.legend()


```
